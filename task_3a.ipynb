```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub
path = kagglehub.dataset_download("shivamb/netflix-shows")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(f"{path}/netflix_titles.csv")
df.head()
```

```python
plt.figure()
df['release_year'].hist(bins=20)
plt.title("Distribution of Release Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()

print("Insight: Most Netflix content was released after 2010.")
```

```python
plt.figure()
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows")
plt.show()

print("Insight: Netflix has more Movies than TV Shows.")
```

```python
plt.figure()
sns.boxplot(x=df['release_year'])
plt.title("Outliers in Release Year")
plt.show()

print("Insight: Very old movies appear as outliers.")
```

```python
plt.figure()
sns.heatmap(df.select_dtypes(include=['int64']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

print("Insight: No strong correlation between numerical features.")
```

```python
important_features = ['type', 'release_year', 'rating', 'duration', 'country']
print("Important features for prediction:", important_features)
```

```python
print("""
Summary:
- Netflix content increased rapidly after 2010.
- Movies dominate the platform.
- Older movies are rare and appear as outliers.
- Numerical features show weak correlation.
""")
```
